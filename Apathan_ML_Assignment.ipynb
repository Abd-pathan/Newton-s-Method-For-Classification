## Newton's Method

```python
#Calculating sigmoidal function

def moid(z):
    return 1/(1+np.exp(-z))

def c_fun(X, y, th):
    h = moid(np.dot(X, th))
    cost = y * np.log(h) + (1 - y) * np.log(1 - h)
    return -cost.mean()
```

```python
from matplotlib import pyplot as plt
import math, random
from sklearn.model_selection import train_test_split
for i in range(10):
    Xtrain, Xtest, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=i)

    X = np.array(Xtrain)
    y = np.array(y_train)
    y = y.reshape(y.size, 1)

    th = np.zeros((X.shape[1], 1))


    c1 = c_fun(X, y, th)
    co_list = [c1]
    m, n = X.shape
    while True:
        h = moid(np.dot(X, th))
        gradient = np.dot(X.T, (h - y)) / y.size
        diag = np.multiply(h, (1 - h)) * np.identity(m)
        hessian = (1 / m) * np.dot(np.dot(X.T, diag), X)
        th = th - np.dot(np.linalg.inv(hessian), gradient)
        c2 = c_fun(X, y, th)
        co_list.append(c2)

        if c2 >= c1 or math.isnan(c2):
            break
        else:
            c1 = c2


    Xtest = np.array(Xtest)
    y_test = np.array(y_test)

    r_test, col_test = Xtest.shape;
    rowy_test = y_test.shape;

    p = []
    for j in range(r_test):
        if moid(np.dot(Xtest[j],th)) >= 0.5:
            p.append(1);
        else:
            p.append(0);

    corr_clf = 0;
    inco_corr_clf= 0;
    error = []
    ERROR = []
    CLF_ACC = []
    for j in range(r_test):
        error.append((p[j] - y_test[j])**2)
        if p[j] == y_test[j]:
            corr_clf = corr_clf + 1
        else:
            inco_corr_clf = inco_corr_clf + 1

    ERROR.append((1/2) * sum(error))
    CLF_ACC.append(corr_clf/rowy_test[0])

    
    
    print("CLASSIFICATION ACCURACY: ", CLF_ACC[-1])
    print("ERROR: ", ERROR[-1])

    plt.plot(co_list)
    plt.xlabel('iterations')
    plt.ylabel('error')
    plt.show()
```
